```python
df=pd.read_csv("/content/HumanBodyPerformance.csv")
df
```

**import matplotlib **

```python
import matplotlib.pyplot as plt
```

# Scatter plot

```python
plt.scatter(df['age'],df["height_cm"],c="red")
plt.title("scatter plot" , fontsize=15)
plt.xlabel("age")
plt.ylabel("height_cm")
plt.show
```

```python
plt.scatter(df['gender'],df["height_cm"],c="blue")
plt.title("scatter plot" , fontsize=15)
plt.xlabel("gender")
plt.ylabel("height_cm")
plt.show
```

```python
plt.scatter(df['age'],df["weight_kg"],c="y")
plt.title("scatter plot" , fontsize=15)

plt.xlabel("age")
plt.ylabel("weight_kg")
plt.show
```

```python
plt.scatter(df['gender'],df["weight_kg"],c="green")
plt.title("scatter plot" , fontsize=15)
plt.xlabel("gender")
plt.ylabel("weight_kg")
plt.show
```

# Histogram plot

```python
plt.hist(df["age"],color="red",edgecolor="black", bins=15)
plt.title("histogram" , fontsize=15)
plt.xlabel("age")
plt.show()
```

```python
plt.hist(df["gender"],color="b",edgecolor="black")
plt.title("histogram" , fontsize=15)
plt.xlabel("gender")
plt.show()
```

```python
plt.hist(df["height_cm"],color="y",edgecolor="black")
plt.title("histogram" , fontsize=15)
plt.xlabel("height_cm")
plt.show()
```

# Bar plot

```python
plt.bar(df["gender"],df["height_cm"],color="g",width=0.1,align="center")
plt.title("bar plot" , fontsize=15)
plt.xlabel("gender")
plt.ylabel("height_cm")
plt.show()
```

```python
plt.bar(df["age"],df["height_cm"],color="b",width=0.5,align="center")
plt.title("bar plot" , fontsize=15)
plt.xlabel("age")
plt.ylabel("height_cm")
plt.show()
```

```python
plt.bar(df["gender"],df["weight_kg"],color="g",width=0.5,align="center")
plt.title("bar plot" , fontsize=15)
plt.xlabel("gender")
plt.ylabel("weight_kg")
plt.show()
```
